# **FINE TUNE USING LORA**

##Tokenizing

```python
import torch
# Load the tokenizer for your model


# Ensure the padding token is defined
if tokenizer.pad_token is None:
    tokenizer.add_special_tokens({'pad_token': '[PAD]'})



# Break long texts into overlapping chunks

MAX_LEN = 256     # start here; you can try 384 later
OVERLAP = 64

def sliding_window_tokenize(text, tokenizer, max_length=MAX_LEN, overlap=OVERLAP):
    # tokenize to *ids* directly and slice ids; this avoids token->string->token roundtrips
    ids = tokenizer.encode(text, add_special_tokens=False)
    chunks = []
    step = max_length - overlap
    for i in range(0, len(ids), step):
        chunks.append(ids[i:i+max_length])
    return chunks


# Generate summaries for long texts (handles multiple chunks)
def generate_summary(model, tokenizer, input_text, max_length=100, overlap=10):
    chunks = sliding_window_tokenize(input_text, tokenizer, max_length=max_length, overlap=overlap)
    generated_summaries = []

    # Generate summary for each chunk
    for chunk in chunks:
        # Convert tokenized chunks back into text
        chunk_text = tokenizer.convert_tokens_to_string(chunk)

        # Tokenize the chunk text and move to GPU
        inputs = tokenizer(chunk_text, return_tensors="pt", padding="max_length", truncation=True, max_length= MAX_LEN).to("cuda")

        with torch.no_grad():
            outputs = model.generate(inputs['input_ids'], max_new_tokens=200)

        summary_chunk = tokenizer.decode(outputs[0], skip_special_tokens=True)
        generated_summaries.append(summary_chunk)

    # Combine summaries from all chunks into the final summary
    final_summary = " ".join(generated_summaries)
    return final_summary

# Correct Tokenization Function
def tokenize_function(examples):
    input_ids_list = []
    attention_masks_list = []

    for article in examples["article"]:
        # break a long article into <=MAX_LEN chunks of token IDs
        chunks = sliding_window_tokenize(article, tokenizer, max_length=MAX_LEN, overlap=OVERLAP)
        for ids in chunks:
            # pad to MAX_LEN using the tokenizer’s pad_token_id
            attn = [1]*len(ids)
            if len(ids) < MAX_LEN:
                pad_len = MAX_LEN - len(ids)
                ids  = ids + [tokenizer.pad_token_id]*pad_len
                attn = attn + [0]*pad_len

            input_ids_list.append(ids)
            attention_masks_list.append(attn)

    return {"input_ids": input_ids_list, "attention_mask": attention_masks_list}


# Example to load your dataset (adjust paths accordingly)
from datasets import load_dataset
from google.colab import drive
drive.mount('/content/drive')

dataset = load_dataset('json', data_files={"train": "/content/drive/MyDrive/Thesis_Datasets/splits/bwsd_prothomalo/bwsd_prothomalo_clean_train.jsonl",
                                            "validation": "/content/drive/MyDrive/Thesis_Datasets/splits/bwsd_prothomalo/bwsd_prothomalo_clean_val.jsonl",
                                            "test": "/content/drive/MyDrive/Thesis_Datasets/splits/bwsd_prothomalo/bwsd_prothomalo_clean_test.jsonl"})

# Tokenize the dataset using the tokenize function defined earlier
train_dataset = dataset["train"].map(tokenize_function, batched=True, remove_columns=dataset["train"].column_names)
valid_dataset = dataset["validation"].map(tokenize_function, batched=True, remove_columns=dataset["validation"].column_names)
test_dataset  = dataset["test"].map(tokenize_function, batched=True, remove_columns=dataset["test"].column_names)

# make them PyTorch-ready *after* map
train_dataset.set_format(type="torch", columns=["input_ids","attention_mask"])
valid_dataset.set_format(type="torch", columns=["input_ids","attention_mask"])
test_dataset.set_format(type="torch",  columns=["input_ids","attention_mask"])
```

##Lora wights

```python
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training

# (Optional, but recommended for k-bit fine-tuning)
model = prepare_model_for_kbit_training(model)

# LoRA config — small rank keeps VRAM low on a 16 GB GPU
lora_config = LoraConfig(
    r=16,                      # try 8; increase to 16 later if VRAM allows
    lora_alpha=16,
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=[
        "q_proj", "k_proj", "v_proj", "o_proj",  # attention proj layers
        "gate_proj", "up_proj", "down_proj"      # MLP proj layers (LLaMA-style)
    ]
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()  # sanity check
```

##Model Train

```python
# Save checkpoints to Colab's local `/content` (instead of Google Drive)
OUTPUT_DIR = "/content/llama3_prothomalo_checkpoints"  # Save checkpoints to local `/content` folder

# Enable gradient checkpointing
import transformers
print("Transformers version:", transformers.__version__)

model.gradient_checkpointing_enable()
model.config.use_cache = False  # required with checkpointing

from transformers import DataCollatorForLanguageModeling
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

from transformers import TrainingArguments
from transformers import Trainer

# Training Arguments with the desired configurations
training_args = TrainingArguments(
    output_dir=OUTPUT_DIR,                    # Save to /content/llama3_prothomalo_checkpoints
    num_train_epochs=1,
    per_device_train_batch_size=16,           # keep small batch size to save VRAM
    per_device_eval_batch_size=16,
    gradient_accumulation_steps=16,           # effective batch size = 32
    learning_rate=2e-4,
    lr_scheduler_type="cosine",
    warmup_ratio=0.03,
    logging_steps=10,
    save_strategy="steps",                    # save checkpoint after each `save_steps`
    save_steps=70,                           # save every 250 steps
    save_total_limit=2,                       # keep only the last 2 checkpoints
    fp16=True,                                # mixed precision for memory efficiency
    optim="adamw_bnb_8bit",
    report_to=[]
)

# Callback for auto-downloading checkpoints every `save_steps`
from transformers import TrainerCallback
import shutil
from google.colab import files
import os

class AutoDownloadCallback(TrainerCallback):
    def __init__(self, save_dir="/content/checkpoints", download_every=70):
        self.save_dir = save_dir
        self.download_every = download_every

    def on_step_end(self, args, state, control, **kwargs):
        # Trigger every N steps
        if state.global_step % self.download_every == 0 and state.global_step > 0:
            zip_path = f"{self.save_dir}_step{state.global_step}.zip"
            # Zip the checkpoint folder
            shutil.make_archive(zip_path.replace(".zip", ""), 'zip', self.save_dir)
            print(f"\n📦 Zipped checkpoint at step {state.global_step}: {zip_path}")
            try:
                files.download(zip_path)  # Auto-download to your PC
            except Exception as e:
                print("⚠️ Download failed (likely Colab disconnect). Zip is still saved:", zip_path)


# Trainer setup with AutoDownloadCallback
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=valid_dataset,
    data_collator=data_collator,
    callbacks=[AutoDownloadCallback(save_dir=OUTPUT_DIR, download_every=70)]  # Save and download after 250 steps
)

import gc, torch
from transformers.trainer_utils import get_last_checkpoint

gc.collect(); torch.cuda.empty_cache()

# Check if checkpoints exist in the specified directory
last_ckpt = get_last_checkpoint(OUTPUT_DIR)
if last_ckpt is not None:
    print(f"Resuming from checkpoint: {last_ckpt}")
    train_result = trainer.train(resume_from_checkpoint=last_ckpt)
else:
    print("No checkpoint found, starting fresh.")
    train_result = trainer.train()

# Evaluate
print(trainer.evaluate(eval_dataset=valid_dataset))

gc.collect()
torch.cuda.empty_cache()

# Example generation
example = valid_dataset[0]
inputs = {
    "input_ids": example["input_ids"].unsqueeze(0),
    "attention_mask": example["attention_mask"].unsqueeze(0),
}

with torch.no_grad():
    gen = model.generate(
        inputs["input_ids"].to(model.device),
        attention_mask=inputs["attention_mask"].to(model.device),
        max_new_tokens=200,
        do_sample=False
    )

print(tokenizer.decode(gen[0], skip_special_tokens=True))
```

##Generate Summary

```python
import json
import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from tqdm.auto import tqdm

# Path to your JSONL validation dataset
dataset_path = '/content/bwsd_prothomalo_clean_val.jsonl'

# Load the dataset
def load_dataset(file_path):
    dataset = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            obj = json.loads(line)
            dataset.append(obj)
    return dataset

data = load_dataset(dataset_path)

# Limit to the first 50 entries
# subset_size = 50
# data = data[:subset_size]

articles   = [d['article'] for d in data]
references = [d['summary'] for d in data]

# Move model to GPU/CPU
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model.to(device)

# Generation parameters
batch_size        = 20
max_input_length  = 512      # input truncation length
max_new_tokens    = 128      # how many tokens to generate in the summary
num_beams         = 4



# Generate summaries and store in `predictions`
predictions = []
for i in tqdm(range(0, len(articles), batch_size), desc="Generating"):
    batch = articles[i:i+batch_size]
    inputs = tokenizer(
        batch,
        max_length=max_input_length,
        truncation=True,
        padding=True,
        return_tensors='pt'
    ).to(device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,  # specify new tokens to generate
            num_beams=num_beams,
            no_repeat_ngram_size=3,
            early_stopping=True,
            do_sample=False
        )

    decoded = tokenizer.batch_decode(
        outputs,
        skip_special_tokens=True,
        clean_up_tokenization_spaces=True
    )
    predictions.extend([s.strip() for s in decoded])

print(f"Generated {len(predictions)} summaries.")
# At this point `predictions` and `references` can be used for evaluation.
```

##Evaluate with rouge, bert and bleu

```python
import numpy as np
import evaluate

# Install missing dependencies if needed
# !pip install rouge_score sacrebleu bert-score evaluate

# Compute ROUGE (1/2/L) and BLEU
rouge = evaluate.load('rouge')
rouge_scores = rouge.compute(
    predictions=predictions,
    references=references,
    rouge_types=['rouge1', 'rouge2', 'rougeL'],
    use_aggregator=True
)

bleu = evaluate.load('bleu')
bleu_result = bleu.compute(
    predictions=predictions,
    references=[[ref] for ref in references]
)

# Compute BERTScore WITHOUT baseline rescaling to avoid the error
bertscore = evaluate.load('bertscore')
bs_result = bertscore.compute(
    predictions=predictions,
    references=references,
    lang='bn',
    rescale_with_baseline=False  # disable baseline rescaling:contentReference[oaicite:3]{index=3}
)
bertscore_f1        = float(np.mean(bs_result['f1']))
bertscore_precision = float(np.mean(bs_result['precision']))
bertscore_recall    = float(np.mean(bs_result['recall']))

# Save results to metrics.txt
with open('Llama3-prothomalo-metrics.txt', 'w', encoding='utf-8') as f:
    f.write(f"ROUGE-1: {rouge_scores['rouge1']:.4f}\n")
    f.write(f"ROUGE-2: {rouge_scores['rouge2']:.4f}\n")
    f.write(f"ROUGE-L: {rouge_scores['rougeL']:.4f}\n")
    f.write(f"BLEU:    {bleu_result['bleu']:.4f}\n")
    f.write(f"BERTScore F1: {bertscore_f1:.4f}\n")
    f.write(f"BERTScore Precision: {bertscore_precision:.4f}\n")
    f.write(f"BERTScore Recall: {bertscore_recall:.4f}\n")


from google.colab import files
files.download("Llama3-prothomalo-metrics.txt")
print("Metrics saved to Llama3-prothomalo-metrics.txt")
```
